```python
import pandas as pd      #data handling
import numpy as np        #numerical operations
import matplotlib.pyplot as plt      #data visualization
import seaborn as sns        #advanced visualizations
from sklearn.model_selection import train_test_split    #for ML tasks

# set dataframe 
bank_df=pd.read_csv("bank.csv")
nba_df=pd.read_csv("all_seasons.csv")

# print the head of dataframe
print(nba_df.head())
print(bank_df.head())
```

```python
print("bank marketing datasets shape: ", bank_df.shape)
print("NBA Players dataset shape: ", nba_df.shape)
```

```python
# tells the datatype of each column(imp for feature selection)
print("Bank Dataset columns: \n",nba_df.dtypes)
print("NBA players Dataset columns: \n",bank_df.dtypes)
```

```python
# to check missing values(sum() used to find the total count of the missing values)
print("Missing values in bank dataset: \n,",bank_df.isnull().sum())
print("Missing values in NBA Players dataset: \n,",nba_df.isnull().sum())
```

```python
# to generate the statistical summary of the datset(we use describe())har numeric column ke mean,min,max,std deviation wagera dikhayega
print("Bank dataset summary: \n",bank_df.describe())
print("NBA Players dataset summary: \n",nba_df.describe())
```
